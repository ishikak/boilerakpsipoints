```python
df = pd.read_excel("Book1.xlsx")
pd.set_option("display.max_rows", None, "display.max_columns", None)
ending = pd.DataFrame(columns = ["First Name", "Last Name", "Year", "Good Standing", "Checkpoint", 
                                 "Points Goal (In Checkpoint)", "Points Earned(In Checkpoint)",
                                 "Points Difference", "Event", "Event Date", "Contributive", "Points Recieved"])

subset = df.iloc[5:63, np.r_[0:5,31, 32, 6:19,21,23:28]].reset_index().drop(['index'], axis=1)
events = df.iloc[0:4,np.r_[6:19,21,23:28]]
events = events.rename(columns={x:y for x,y in zip(events.columns,range(0,len(events.columns)))})
for m, row in subset.iterrows(): 
    for i in range(19):
        ending = ending.append({'First Name': subset['~57 Active Brothers'].values[m], 'Last Name':subset['Point System'].values[m],
                                'Year': subset['Unnamed: 0'].values[m] ,'Checkpoint': 1,
                                'Points Goal (In Checkpoint)': float(subset['Unnamed: 32'].values[m]),
                                'Points Earned(In Checkpoint)': float(subset['Unnamed: 31'].values[m]), 
                                'Points Difference':float((subset['Unnamed: 31'].values[m])-(subset['Unnamed: 32'].values[m])), 
                                'Good Standing': subset['Unnamed: 4'].values[m]}, ignore_index=True)
        
#.values[m]
```

```python
list2 = events.T[0].to_list() 
l = []
for i in range(58): 
    for m in list2 : 
        l.append(m) 
ending['Event'] = l      
        
list1 = events.T[1].to_list()
l = []
for i in range(58): 
    for m in list1 : 
        l.append(m) 
ending['Event Date'] = l


list3 = events.T[2].to_list()
l = []
for i in range(58): 
    for m in list3: 
        l.append(m) 
ending['Contributive'] = l
#ending.head(19)
len(list2)
```

```python
points = df.iloc[5:63,np.r_[6:19,21,23:28]]
original_list = df.iloc[3:4,np.r_[6:19,21,23:28]].values.tolist() # point values for each event, list in list 
flattened = [val for sublist in original_list for val in sublist] # flattened values, all in one list 

#def pointschange(df)
for column in points.columns:
    ind = points.columns.get_loc(column)
    point_val = flattened[ind]
    list_of_vals = points[column].tolist()
    for index, item in enumerate(list_of_vals): 
        if item == "Yes":
            list_of_vals[index] = point_val
        else: 
            list_of_vals[index] = 0
    points[column] = list_of_vals

```
